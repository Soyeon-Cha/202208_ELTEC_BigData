```python
# 합, 평균 열 만들기
df = pd.read_excel("성적 처리.xlsx", sheet_name = 'Sheet1')
df
```

```python
# df['합'] = df['국어'] + df['영어'] + df['수학'] + df['사회'] + df['과학']
# df['평균'] = df['합'] / 5
# df

df['합'] = df.iloc[:, 2:].sum(axis = 1)
df['평균'] = df['합'] / 5
df
```

```python
df
```

```python
df.groupby("반").sum()
```

```python
df.groupby("반").mean()
```

```python
# 순위

# average : 동점은 평균으로 순위 부여
# 3위가 2개이면, 3.5로 순위를 부여하고 다음 순위는 5위

df['순위'] = df['평균'].rank(method = 'average', ascending=False)
df
```

```python
# min : 동점은 취소 순위 부여
# 3위가 2개이면, 둘다 3위로 부여, 다음 순위는 5등

df['순위'] = df['평균'].rank(method = 'min', ascending=False)
df
```

```python
# average : 동점은 최대값으로 순위 부여
# 3위가 2개이면, 4로 순위를 부여하고 다음 순위는 5위

df['순위'] = df['평균'].rank(method = 'max', ascending=False)
df
```

```python
# average : 동점은 평균으로 순위 부여
# 3위가 2개이면, 처음 발견한 3등 3등, 4등

df['순위'] = df['평균'].rank(method = 'first', ascending=False)
df
```

```python
# average : 동점은 평균으로 순위 부여
# 3위가 2개이면, min이랑 같다. 둘다 3위, 그 다음 순위 4등

df['순위'] = df['평균'].rank(method = 'dense', ascending=False)
df
```

```python
# 데이터 크롤링 : 인터넷에서 데이터를 수집

# HTML : 파이썬으로 HTML을 다루는 기능이 없을까?
# 페이지를 요청하는 기능이 없을까?

import requests  # 페이지 요청
from bs4 import BeautifulSoup # 파이썬으로 HTML을 다루는 기능
```

```python
url = requests.get("https://search.daum.net/search?w=tot&DA=UME&t__nil_searchbox=suggest&sug=&sugo=16&sq=%EB%A1%9C%EB%98%90&o=4&q=%EB%A1%9C%EB%98%90")

# 200 : 정상 / 400번대 : 존재하지않는페이지 // 500번대 : 로그인
# 어떻게 데이터를 쓰냐에 불법

# 전체 데이터 가져오면 : 크롤링
# 일부분 가져오면 : 스크래핑

html = BeautifulSoup(url.text) # url.text를 html로 변환해줘!
html
```

```python
# span 태그 (class 속성값 f_red)

current = int(html.find("span", class_ = 'f_red').text.replace("회", ""))
current
```
