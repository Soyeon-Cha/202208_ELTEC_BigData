```python
# 클래스 : 변수나 함수들을 모아놓은 주머니
# 붕어빵을 만드는 틀 : 객체를 만들어 내는 것

# class 클래스명

class Bank:

    # 초기화함수 : 객체를 생성할 때, 자동으로 한번 실행되는 함수
    def __init__(self):
        print("가입을 축하합니다.")
        self.money = 0

    def deposit(self, a):
        self.money += a
```

```python
person1 = Bank() # person1은 Bank에서 제공하는 모든 서비스를 사용할 수 있는 권한
```

```python
person2 = Bank()
```

```python
person1.deposit(30000)
```

```python
person2.deposit(50000)
```
